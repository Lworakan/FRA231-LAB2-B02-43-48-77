```python
# plot graph of motor characteristics
# Speed vs Torque
# y intercept is 249.91 rad/s
# x intercept is 0.2430 Nm

import matplotlib.pyplot as plt
import numpy

x_intercept = 0.2430 # Nm
y_intercept = 249.91 # rad/s
slope = - y_intercept / x_intercept

# plot from torque = 0 to 0.2430 Nm

torque = numpy.linspace(0, 0.2430, 100)
speed = slope * torque + y_intercept
print('slope =', slope)

plt.plot(torque, speed)
plt.title('Motor Characteristics - Speed vs Load Torque')
plt.xlabel('Torque (Nm)')
plt.ylabel('Speed (rad/s)')
plt.grid()
plt.show()

```
